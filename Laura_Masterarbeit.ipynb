# Kollerationsmatrix

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

csv_korrelation = "https://raw.githubusercontent.com/brunryan/masterarbeit_laura/main/Daten_roh.csv"

# Schritt 1: CSV-Datei einlesen
df = pd.read_csv(csv_korrelation, sep=';', decimal=',')

df.describe().T
```

```python
# Überprüfen der Daten
print(df.head())

# Berechnen der Korrelationsmatrix
corr_matrix = df.corr()

# Korrelogramm erstellen
plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar_kws={"shrink": .8}, annot_kws={"size": 8})
plt.title('Korrelogramm der Variablen')
plt.show()
```

```python

# Berechnen der Korrelationsmatrix
corr_matrix = df.corr()

# Maske für die obere Dreiecksmatrix erstellen
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Korrelogramm erstellen
plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar_kws={"shrink": .8}, annot_kws={"size": 8})
plt.title('Korrelogramm der Variablen')
plt.show()
```
