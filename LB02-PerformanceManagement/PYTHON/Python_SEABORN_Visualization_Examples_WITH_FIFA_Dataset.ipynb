# Load the FIFA data

In this notebook, we'll work with a dataset of **historical FIFA rankings** for six countries:   
+ Argentina (ARG),  
+ Brazil (BRA),
+ Spain (ESP),
+ France (FRA),
+ Germany (GER), and   
+ Italy (ITA).  

The dataset is stored as a **CSV file** (short for *comma-separated values* file). Opening the CSV file in Excel shows a row for each date, along with a column for each country.

![tut0_fifa_head](https://i.imgur.com/W0E7GjV.png)

To load the data into the notebook, we'll use two distinct steps, implemented in the code cell below as follows:   
+ begin by specifying the location (or *filepath*) where the dataset can be accessed, and then   
+ use the filepath to load the contents of the dataset into the notebook.

```python
# Path of the file to read data from github repository
fifa_filepath = "https://raw.githubusercontent.com/sawubona-repo/BINA-FS24-WORK/main/LB02-PerformanceManagement/DATA/DATA_fifa.csv"

# Read the file into a (pandas dataframe) variable fifa_data
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
```

> Note that the code cell above has **four** different lines.

# B) Plot the data

In this course, you'll learn about many different plot types.  In many cases, you'll only need one line of code to make a chart!

For a sneak peek at what you'll learn, check out the code below that generates a **line chart**.

```python
# Set the width and height of the figure
plt.figure(figsize=(15,5))

# Line chart showing how FIFA rankings evolved over time
sns.lineplot(data=fifa_data, markers=False, dashes=False, legend='auto')
```

# C) Beautify the plot

You can modify and add multiple elements of a digramm/plot using all the features and functions of the *matplotlib* library underlying the easy to use **seaborn** visualization library!


```python
# Plot the data with additional Matplotlib settings
plt.figure(figsize=(15,5))

# set fonty style for title and legends
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

# Line chart showing how FIFA rankings evolved over time
sns.lineplot(data=fifa_data, markers=False, dashes=False, legend='auto')

# plot a diagram TITLE
plt.title("Historical FIFA Ranking", loc = 'left', fontdict = font1)

# add AXIS LABEL text
plt.xlabel("Year", fontdict = font2)
plt.ylabel("Ranking", fontdict = font2)

# add and position a LEGEND BOX
plt.legend(ncol=6, loc='upper center')

# save the plot as PNG file
plt.savefig("Historical-FIFA-Ranking.png")

plt.show()
```

# D) Explore the dataset visually

You can **select columns** or **filter rows** using different visualization diagram types to explore the original dataset

### D.1) Show only **selected** countries

```python
# compare ARG vs FRA
sample_data = fifa_data[['ARG','FRA']]

sample_data.head()
```

```python
# set plot figure size
plt.figure(figsize=(15,5))

# set fonty style for title and legends
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':10}

# Line chart showing how FIFA rankings evolved over time
sns.lineplot(data=sample_data, markers=False, dashes=False, legend='auto')

# plot a diagram TITLE
plt.title("Historical FIFA Ranking", loc = 'center', fontdict = font1)
plt.suptitle("Line Diagram", fontdict = font1)

# add axis label text
plt.xlabel("Year", fontdict = font2)
plt.ylabel("Ranking", fontdict = font2)

# add and position a legend box
plt.legend(ncol=3, loc='upper center')

plt.show()
```

### D.2) Compare **pairs of countries** using scatterplot diagram

```python
# set plot figure size
plt.figure(figsize=(7,7))

# Scatter (correlation) chart showing how FIFA rankings evolved over time
sns.scatterplot(data=fifa_data, x="ARG", y="FRA", markers=True, color="darkred", s=100, alpha=0.25)

# plot a diagram title
plt.title("Scatter Diagram: ARG vs FRA", loc = 'center', fontdict = font1)

plt.show()
```

```python
# pairwise Scatter diagram
fig = sns.pairplot(data=fifa_data)

# plot a diagram title
plt.title("Pairwise Scatter Diagram", loc = 'center', fontdict = font1)

plt.show()
```

### D.3) Plot the key descriptive statistics measures using **boxplot** (*Box Whisker*) diagram

```python
# set plot figure size
plt.figure(figsize=(5,5))
# multiple Box Whisker diagram
sns.boxplot(data=fifa_data, order=['ARG','FRA'], orient ='v')

# plot a diagram title
plt.title("Box Whisker Diagram", loc = 'center', fontdict = font1)

plt.show()
```

```python
print(fifa_data.columns.tolist())
```

```python
# set plot figure size
plt.figure(figsize=(10,5))

# multiple Box Whisker diagram
sns.boxplot(data=fifa_data, order=fifa_data.columns.tolist(), orient='h')

# plot a diagram title
plt.title("Box Whisker Diagram", loc = 'left', fontdict = font1)

plt.show()
```

### D.4) Plot the distribution or density **histplot** (*Histogram*) diagram

```python
# set plot figure size
plt.figure(figsize=(10,5))

# Histogram diagram (playing with different bin sizes)
sns.histplot(data=fifa_data, x="FRA", bins=10)

# plot a diagram title
plt.title("Histogram", loc = 'left', fontdict = font1)

# control y axe limits
plt.ylim(0, 150)

plt.show()
```

```python
# set plot figure size
plt.figure(figsize=(10,5))

# stacked Histogram diagram
sns.histplot(data=fifa_data)

# plot a diagram title
plt.title("Stacked Histogram", loc = 'left', fontdict = font1)

plt.show()
```

```python
# set plot figure size
plt.figure(figsize=(10,5))

# Density diagram
sns.kdeplot(data=fifa_data, x="FRA")

# plot a diagram title
plt.title("Density Diagram", loc = 'left', fontdict = font1)

plt.show()
```

```python
# set plot figure size
plt.figure(figsize=(10,5))

# multiple Density diagrams
fig = sns.kdeplot(data=fifa_data, x="ARG", fill=True, color="b")
fig = sns.kdeplot(data=fifa_data, x="FRA", fill=True, color="r")
fig = sns.kdeplot(data=fifa_data, x="GER", fill=True, color="y")

# plot a diagram title
plt.title("Density Diagram", loc = 'left', fontdict = font1)

plt.legend("AFG", loc='upper right')

plt.ylim(0, 0.20)

plt.show()
```

### D.5) Plot a **connected scatterplot** diagram

```python
# selecting rows based on position
result_data = fifa_data[0:20]

result_data.head()
```

```python
# set plot figure size
plt.figure(figsize=(7,7))

# connected Scatter diagram
plt.plot("ARG", "FRA", data=result_data, linestyle='dotted', marker='o', alpha=0.5)

# plot a diagram title
plt.title("Connected Scatter Diagram", loc = 'left', fontdict = font1)

# add axis label text
plt.xlabel("ARG", fontdict = font2)
plt.ylabel("FRA", fontdict = font2)

# control axe limits
plt.ylim(0, 25)
plt.xlim(0, 25)

plt.show()
```
